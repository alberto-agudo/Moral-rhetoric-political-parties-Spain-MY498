```python
import pandas as pd
import pickle
import numpy as np
import os
import re
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
sns.set()
# ---
# ---
# ---
def transform_str_to_probs_list(probabilities):
    """
    Takes a string of probabilities in the following shape: '[0.99, 0.1]'
    and turns them into a list of floating point probs.
    """
    return [float(prob) for prob in probabilities.lstrip("[").rstrip("]").split(", ")]

def distributions_plot(dataset, variable, 
                       title = "", xlabel = None, 
                       **sns_displot_kwargs):
    """
    Creates a distribution plot of a variable showing its differences by party.
    """
    sns.set_style("whitegrid")
    # Set the colors of each party
    parties_palette = {"VOX": "tab:green",
                       "PP": "tab:blue",
                       "CS": "tab:orange",
                       "PSOE": "tab:red",
                       "PODEMOS": "tab:purple"}
    ax = sns.displot(dataset, x = variable, hue = "party", 
                     kind = "kde", palette = parties_palette, **sns_displot_kwargs)
    
    font = {'family' : 'serif'}
    plt.rc('font', **font)
    plt.title(title, 
              size = 18, fontweight = "bold")
    plt.xlabel(xlabel, size = 15)
    plt.ylabel("Density", size = 15)
    plt.xticks(size = 14)
    plt.yticks(size = 14)
    
    # Change legend title and increase size
    new_title = "Party"
    ax._legend.set_title(new_title)
    plt.setp(ax._legend.get_title(), fontsize = 12)
    plt.setp(ax._legend.get_texts(), fontsize = 12)
    plt.show()
# ---
pd.set_option("display.max_colwidth", 150)
```

- Create a dictionary that links each file to its transcription and the probabilities of each word:

```python
trans_dict = {}
for file_id in range(len(file_paths)):
    short_file = re.findall("[0-9]{2,}.mp3", file_paths[file_id])[0]
    trans_dict[short_file] = {}
    
    # If it is one of the manual transcriptions, use its transcription as value: 
    if short_file in manual_trans:
        trans_dict[short_file]['transcription'] = manual_trans[short_file][0]
        trans_dict[short_file]['probabilities'] = [1 for word in range(len(manual_trans[short_file][0]))]
        
    # Files that had an error
    elif short_file in other_transcriptions:
        
        file_info = other_transcriptions[short_file][0][0]
        
        del file_info['start_timestamps']
        del file_info['end_timestamps']
        
        trans_dict[short_file] = file_info
    
    # Otherwise, get only the necessary information
    else:
        file_info = transcriptions[file_id][0]
        
        # Erase unnecessary information:
        del file_info['start_timestamps']
        del file_info['end_timestamps']
        
        trans_dict[short_file] = file_info
```

Create a DataFrame with the transcriptions and probabilities of each audio

```python
# Create a dataframe out of the previous dictionary:
transcriptions_df = pd.DataFrame.from_dict(trans_dict, orient = 'index').reset_index()

# Rename the columns of the dataframe
transcriptions_df = transcriptions_df.rename(columns = {'index': 'audio_file'})

# Change the audio file name, erasing the .mp3 information:
transcriptions_df.audio_file = transcriptions_df.audio_file.apply(lambda x: re.findall("[0-9]{2,}", x)[0])

transcriptions_df
```

### Read and clean the raw dataset:

The probabilities here represent the probability of each character being transcribed correctly.

```python
transcriptions_df = pd.read_csv("../data/raw/transcriptions_df.csv", 
                                dtype = {"status_id": str})

# Identify files without sound:
transcriptions_df[transcriptions_df.probabilities.isnull()]
```

```python
# Eliminate transcriptions without sound:
transcriptions_df = transcriptions_df.loc[transcriptions_df.probabilities.notnull(), :]
transcriptions_df = transcriptions_df.reset_index(drop = True)

# Drop duplicated audios:
transcriptions_df = transcriptions_df.drop_duplicates("transcription").reset_index(drop = True)

# Clean probabilities (they are in the shape of a string, we need to turn them
# into a list of floating point numbers)
transcriptions_df.probabilities = transcriptions_df.probabilities.apply(transform_str_to_probs_list)

# Generate additional features, such as the length of the transcription in words and chars:
transcriptions_df["n_words"] = transcriptions_df.transcription.apply(lambda x: len(x.split(" ")))
transcriptions_df["n_chars"] = transcriptions_df.transcription.apply(len)

# Get aggregate measures of probabilities of correct transcriptions to filter out bad quality audios
transcriptions_df["mean_prob"] = transcriptions_df.probabilities.apply(np.mean)
transcriptions_df["first_quartile"] = transcriptions_df.probabilities.apply(lambda x: np.quantile(x, q = 0.25))
transcriptions_df["median_prob"] = transcriptions_df.probabilities.apply(np.median)
transcriptions_df["third_quartile"] = transcriptions_df.probabilities.apply(lambda x: np.quantile(x, q = 0.75))
transcriptions_df["std_prob"] = transcriptions_df.probabilities.apply(np.std)
```

```python
# Join both tables:
tweets_df = pd.read_csv("../data/clean/tweets_df.csv", 
                        usecols = ["party", "screen_name", "status_id"],
                        dtype = {"status_id": str})

audios_merge = transcriptions_df.merge(tweets_df, on = "status_id")
```

### ELIMINATE NOISY TRANSCRIPTIONS

Given that the transcriptions stem from an AI model, some of them are noisy when the audio quality was bad or if the speakers in the audio spoke too fast. Fortunately, we have the probabilities of each word being transcribed correctly, and we can get a distribution of audio quality.

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 0.2}
distributions_plot(audios_merge, "mean_prob",
                   "Mean probability of correct transcription: Distributions per party", 
                   "Mean probability of correct transcription", **plot_kwargs)
```

It seems like the distribution of probabilities of correct transcription per audio (i.e., the mean probability of the correct transcription per character) lies above 0.9. The party with the cleanest audios is the conservative party (PP), followed by CS and Podemos. Finally, both PSOE (the incumbent party) and VOX (far-right) have the dirtiest audios distribution. Let us check this in terms of the medians:

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 0.1}
distributions_plot(audios_merge, "median_prob",
                   "Median probability of correct transcription: Distributions per party", 
                   "Median probability of correct transcription", **plot_kwargs)
```

Here, as expected, the picture is even more extreme as before, and the probabilities are way closer and centered around 1. Nevertheless, this is not as interesting, because the median probabilities are less sensitive to outliers (which are precisely the noise that we want to remove). Let us check this distribution at the 1st quantile of each probability array:

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 0.1}
distributions_plot(audios_merge, "first_quartile",
                   "First quartile of correct transcription: Distributions per party", 
                   "First quartile of correct transcription", **plot_kwargs)
```

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 3}
distributions_plot(audios_merge, "std_prob", 
                   "Correct transcription probabilities standard deviation: Distributions per party", 
                   "Standard deviation of the probability of correct transcription", **plot_kwargs)
```

Ideally, we would not want the probabilities to deviate too much (over 0.1 might be a problem), especially because a probability below 90% is not too reliable. Let us check how much do audios above 93.5% deviate:

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 1}
distributions_plot(audios_merge[audios_merge.mean_prob > 0.935], "std_prob", 
                   "Correct transcription probabilities standard deviation: Distributions per party above 93.5% mean probability", 
                   "Standard deviation of the probability of correct transcription", **plot_kwargs)
```

Even though the standard deviation distributions have slightly shifted to the left, there is not a big difference, so we will not be able to control the standard deviation of probabilities. Let us take a sample of 30 transcriptions with a mean probability above 90% and consider whether they are reasonable.

```python
plot_kwargs = {"height": 8, "aspect": 1.75, "bw_adjust": 0.75}
distributions_plot(audios_merge, "n_words", 
                   "Distribution of words per audio", 
                   "Number of words in each audio", **plot_kwargs)
```

Audios tend to have around 200 words, but there is high diversity in that sense.

```python
np.random.seed(1)
sample = audios_merge.loc[(audios_merge.mean_prob > 0.935)].sample(30)
for row in sample.index:
    print("Party", sample.loc[row, "party"])
    print("File:", sample.loc[row, "status_id"])
    print("Transcription:", sample.loc[row, "transcription"])
    print("Mean probability:", sample.loc[row, "mean_prob"], "\n")
```

Audios with a mean probability of correct transcription above 93.5% could be considered ok (even though they are not perfect), especially taking into account that some of the audios with 98% mean probability are not much better either. This could be revised in further research with:

    a) Newer and more accurate models
    b) Manual transcription of audios
    c) Google Translate API (which would require the use of a Google Cloud Storage unit for long audios)
